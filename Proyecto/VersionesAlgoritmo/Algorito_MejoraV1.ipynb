```python
import itertools
import time
import pandas as pd

# Iniciar cronómetro
inicio = time.time()

# Paso 1: Preparar dígitos y operadores
digitos = [str(i) for i in range(1, 10)]
operadores = ['+', '-', '*', '/']

# Permutaciones
permutaciones_digitos = itertools.permutations(digitos)  # Generador: 9!
permutaciones_operadores = list(itertools.permutations(operadores))  # Lista: 4!

# Contadores
contador_total = 0
contador_divCero_Cond = 0
contador_divCero_Try = 0
contador_validas = 0
contador_decimales = 0
resultados = []

# Paso 2: Evaluar todas las combinaciones
for d in permutaciones_digitos:
    for o in permutaciones_operadores:
        # Construir expresión como cadena
        expr = d[0]
        for i in range(4):
            expr += o[i] + d[i + 1]
        for j in range(5, 9):
            expr += d[j]


        # 🔍 Mejora 1: filtro por división por cero
        if "/0" in expr:
            contador_divCero_Cond += 1
            continue

        contador_total += 1
        try:
            resultado = eval(expr)
            if isinstance(resultado, int) or resultado.is_integer():
                contador_validas += 1
                resultados.append((expr, int(resultado)))
            else:
              contador_decimales += 1
        except ZeroDivisionError:
            contador_divCero_Try += 1
            continue

# Paso 3: Mostrar resultados
fin = time.time()
duracion = fin - inicio

print(f"Total de expresiones generadas: {contador_total}")
print(f"Expresiones válidas con resultado entero: {contador_validas}")
print(f"Expresiones descartadas por ser decimales: {contador_decimales}")
print(f"Expresiones filtradas debido a división por cero : {contador_divCero_Cond}")
print(f"Expresiones Arrojan error debido a división por cero : {contador_divCero_Try}")
print(f"Duración total: {duracion:.2f} segundos (~{duracion/60:.2f} minutos)")

# Mostrar primeras 10 expresiones válidas
df = pd.DataFrame(resultados, columns=["Expresión", "Resultado"])
print(df.head(10))
```
