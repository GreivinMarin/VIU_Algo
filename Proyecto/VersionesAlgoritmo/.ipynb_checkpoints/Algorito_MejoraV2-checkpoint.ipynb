```python
import itertools
import time
import pandas as pd
import ast

# Iniciar cronómetro
inicio = time.time()

# Paso 1: Preparar dígitos y operadores
digitos = [str(i) for i in range(1, 10)]
operadores = ['+', '-', '*', '/']

# Permutaciones
permutaciones_digitos = itertools.permutations(digitos)  # Generador: 9!
permutaciones_operadores = list(itertools.permutations(operadores))  # Lista: 4!

# Contadores
contador_total = 0
contador_validas = 0
contador_divCero = 0
contador_decimales = 0
contador_ast_error = 0

resultados = []

# Función para construir expresión
def construir_expresion(digs, ops):
    expr = digs[0]
    for i in range(4):
        expr += ops[i] + digs[i + 1]
    for j in range(5, 9):
        expr += digs[j]
    return expr

# Función para validar sintaxis con AST
def es_expresion_valida(expr):
    try:
        ast.parse(expr, mode='eval')
        return True
    except:
        return False

# Paso 2: Evaluar todas las combinaciones
for d in permutaciones_digitos:
    for o in permutaciones_operadores:
        expr = construir_expresion(d, o)
        contador_total += 1

        # Validación sintáctica con AST
        if not es_expresion_valida(expr):
            contador_ast_error += 1
            continue

        try:
            resultado = eval(expr)
            if isinstance(resultado, int) or resultado.is_integer():
                contador_validas += 1
                resultados.append((expr, int(resultado)))
            else:
                contador_decimales += 1
        except ZeroDivisionError:
            contador_divCero += 1
            continue

# Paso 3: Mostrar resultados
fin = time.time()
duracion = fin - inicio

print(f"Total de expresiones generadas: {contador_total}")
print(f"Expresiones válidas con resultado entero: {contador_validas}")
print(f"Expresiones descartadas por ser decimales: {contador_decimales}")
print(f"Errores de división por cero: {contador_divCero}")
print(f"Errores de sintaxis (AST): {contador_ast_error}")
print(f"Duración total: {duracion:.2f} segundos (~{duracion/60:.2f} minutos)")

# Mostrar primeras 10 expresiones válidas
df = pd.DataFrame(resultados, columns=["Expresión", "Resultado"])
print(df.head(10))
```
